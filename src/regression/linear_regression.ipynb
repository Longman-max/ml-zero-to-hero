## Task 1: Predicting Student Grades (Simple Linear Regression)
We will use Python and Scikit-Learn (the industry-standard library for this). We will also use NumPy for data handling and Matplotlib to see our math in action.

Scenario
We want to predict a student's Exam Score based on Hours Studied.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

X = np.array([1, 2, 3, 4, 5, 6, 7]).reshape(-1, 1)
y = np.array([15, 28, 35, 48, 55, 68, 75])


plt.scatter(X, y, color='blue', label="Actual Data")
plt.xlabel("Hours Studied")
plt.ylabel("Exam Score")

plt.show()
```

```python
# train the model

model = LinearRegression()
model.fit(X, y)

print(f"Weight (Slope/w): {model.coef_[0]}")  # this extracts the weight
print(f"Bias (Intercept/b) : {model.intercept_}") # this gets the y-intercept
```

```python
# making prediction

new_hours = np.array([[9.5]])
prediction_score = model.predict(new_hours)

print(f"If you studied for 9.5 hours, you will score: {prediction_score[0]:.2f}")
```

```python
# how to calculate the mean_squared_error

# we first get all the predictions from the original data point
all_predictions = model.predict(X)

# calculate the mse
mse = mean_squared_error(y, all_predictions)

# get the results
print(f"Mean Squared Error: {mse:.2f}")  # to 2 decimal places
print(f"Root Mean Squared Error: {np.sqrt(mse):.2f}")  # to 2 decimal places
```

## Task 2: Real Estate (Multivariate Regression)

The Logic: In the real world, a house price isn't decided by size alone. It depends on size, number of bedrooms, age of the house, etc.Math: Instead of $y = wx + b$, we now have $y = w_1x_1 + w_2x_2 + b$.Visual: Instead of fitting a line in 2D space, the computer is fitting a flat plane in 3D space.The CodeWe will predict Price based on two features: Size (sq ft) and Number of Bedrooms.

```python
import numpy as np
from sklearn.linear_model import LinearRegression

# 1. Prepare Data
# X = [Size in sq ft, Number of Bedrooms]
X = np.array([
    [1400, 2], 
    [1600, 3], 
    [1700, 3], 
    [1875, 4], 
    [1100, 2], 
    [1550, 3], 
    [2350, 4], 
    [2450, 4], 
    [1425, 3], 
    [1700, 4]
])

# y = House Price (in $)
y = np.array([245000, 312000, 320000, 400000, 210000, 305000, 450000, 475000, 300000, 340000])

# 2. Train
model = LinearRegression()
model.fit(X, y)

# 3. Interpret the Brain
# We now have TWO weights (coefficients): One for Size, One for Bedrooms
w1, w2 = model.coef_
b = model.intercept_

print(f"Base Price (Bias): ${b:.2f}")
print(f"Price per Sq Ft (Weight 1): ${w1:.2f}")
print(f"Price per Extra Bedroom (Weight 2): ${w2:.2f}")

# 4. Predict
# Scenario: A 2000 sq ft house with 3 bedrooms
new_house = np.array([[2000, 3]])
prediction = model.predict(new_house)

print(f"Predicted Price for 2000 sq ft, 3 bed: ${prediction[0]:,.2f}")
```
